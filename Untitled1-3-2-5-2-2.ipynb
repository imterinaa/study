## 19 Оптимизация. Задача упаковки кругов

Задача заключается в упаковке в единичный круг $n$ равных друг другу кругов радиуса $r$ так, чтобы r был максимальным.
В функции **cons** вычисляются ограничения для кругов.

    
   
   
    
   

```python
import numpy as np
import random
from scipy.optimize import minimize
import matplotlib.pyplot as plt


def cons(x):
    o = np.array([])
    o = np.append(o, x[0] - 1)
    o = np.append(o, n - x[0])
    for i in range(1, 2*n, 2):
        o = np.append(o, (x[0]-1)**2 - x[i]**2 - x[i+1]**2)
    for i in range(1, 2*n, 2):
        for j in range(i+2, 2*n + 1, 2):
            o = np.append(o, (x[i] - x[j])**2 + (x[i+1] - x[j+1])**2 - 4)
    return o
```

Функция **f**  создаёт массив из $2n+1$ элементов, где $a[0]$ элемент $-$ радиус $R$, а остальные $2n$ элеметов- координаты малых кругов.
    После выполнения метода  **minimize** идёт приведение к первоначальному условию, так как изначально в ограничениях для простоты вычисления взяли упаковку n кругов радиуса 1 при $(R -> min)$. Далее  **f** изображает упаковку и сравнивает решение с значениями из **arr_num**(с решениями, взятых с сайта www.packomania.com)

```python
def f(n):
    xn = 0;
    while (xn < arr_num[n-1]):
        a = np.array([random.uniform(-n,n) for i in range(2*n+1)])
        a[0] = n
        m = minimize(lambda x: x[0], a, constraints=({'type':'ineq','fun': cons}))
        x = m.x / m.x[0]
        x[0] /= m.x[0]
        xn = x[0]
    circle = np.array([])
    fig, ax = plt.subplots()
    plt.xlim(-1.5,1.5)
    plt.ylim(-1.5,1.5)
    ax.set_aspect(1)
    ax.add_artist(plt.Circle((0, 0), 1, color='deeppink'))
    for i in range(1, 2*n+1, 2):
        circle = plt.Circle((x[i], x[i+1]), x[0],color = 'greenyellow')
        ax.add_artist(circle)
    print(n,"circle-error:", abs(xn-arr_num[n-1]))
    return
```
